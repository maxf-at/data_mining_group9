```python
# perform HiSC on dataset

import HiSC

data, labels_true = HiSC.process_csv("../../test_data/basic_ext.csv")

print(data)
print(labels_true)

alpha = 0.006
k = 3
clus_ord = HiSC.HiSC(data, alpha, k, verbose=False, elki=False)

for p in clus_ord:
    print(p)
```

```python
# compute clusters

n_cluster = 3
cluster_order, labels_of_data, labels, threshold = HiSC.get_clusters(clus_ord, n_cluster)

for p in clus_ord:
    print(p)
print(labels_of_data)
print(labels)
print(threshold)
```

```python
# plot reachability

HiSC.reachability_plot(data, cluster_order, labels_true, labels_of_data, labels, threshold)
```
