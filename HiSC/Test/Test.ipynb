```python
# perform HiSC on dataset
import numpy as np
import HiSC

data, labels_true = HiSC.process_csv("../../test_data/basic_ext.csv")

print(data)
print(labels_true)

alpha = 0.006
k = 3
clus_ord = HiSC.HiSC(data, alpha, k, verbose=False, elki=False)

for p in clus_ord:
    print(p)
```

```python
# compute clusters

n_cluster = 2
labels, threshold = HiSC.get_clusters(clus_ord, n_cluster)

print(labels)
print(set(labels))
print(threshold)
```

```python
# plot reachability

HiSC.reachability_plot(data, clus_ord, labels_true, labels, threshold, n_cluster)
```
